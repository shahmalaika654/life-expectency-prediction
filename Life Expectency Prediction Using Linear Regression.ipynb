# TASK #4: CREATE TRAINING AND TESTING DATASET

```python
# Create train and test data

X = life_expectancy_df.drop(columns = ['Life expectancy '])
y = life_expectancy_df[['Life expectancy ']]
```

```python
X
```

```python
y
```

```python
X.shape
```

```python
y.shape
```

```python
# Convert the data type to float32

X = np.array(X).astype('float32')
y = np.array(y).astype('float32')
```

```python
# Only take the numerical variables and scale them
X 
```

```python
# split the data into test and train sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
# Scale the data
from sklearn.preprocessing import StandardScaler

scaler_X = StandardScaler()
X_train = scaler_X.fit_transform(X_train)
X_test = scaler_X.transform(X_test)

scaler_y = StandardScaler()
y_train = scaler_y.fit_transform(y_train)
y_test = scaler_y.transform(y_test)
```

# TASK #5: UNDERSTAND THE THEORY AND INTUITION BEHIND LINEAR REGRESSION

# TASK #6: TRAIN A LINEAR REGRESSION MODEL IN SCIKIT-LEARN

```python
# using linear regression model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, accuracy_score

regresssion_model_sklearn = LinearRegression(fit_intercept = True)
regresssion_model_sklearn.fit(X_train, y_train)
```

```python
regresssion_model_sklearn_accuracy = regresssion_model_sklearn.score(X_test, y_test)
regresssion_model_sklearn_accuracy
```

```python
print('Linear Model Coefficient (m): ', regresssion_model_sklearn.coef_)
print('Linear Model Coefficient (b): ', regresssion_model_sklearn.intercept_)
```

# TASK #7: EVALUATE LR TRAINED MODEL PERFORMANCE 

```python
# Make prediction

y_predict_lr = regresssion_model_sklearn.predict(X_test)
```

```python
#y_predict_lr
```

```python
# Plot the scaled result

plt.plot(y_test, y_predict_lr, "^", color = 'r')
plt.xlabel('Model Predictions')
plt.ylabel('True Values')
```

```python
y_predict_orig = scaler_y.inverse_transform(y_predict_lr)
y_test_orig = scaler_y.inverse_transform(y_test)
```

```python
# Plot the original values

plt.plot(y_test_orig, y_predict_orig, "^", color = 'r')
plt.xlabel('Model Predictions')
plt.ylabel('True Values')
```

```python
# Plot the KPIs

from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from math import sqrt

k = X_test.shape[1]
n = len(X_test)
RMSE_lr = float(format(np.sqrt(mean_squared_error(y_test_orig, y_predict_orig)),'.3f'))
MSE_lr = mean_squared_error(y_test_orig, y_predict_orig)
MAE_lr = mean_absolute_error(y_test_orig, y_predict_orig)
r2_lr = r2_score(y_test_orig, y_predict_orig)
adj_r2_lr = 1-(1-r2_lr)*(n-1)/(n-k-1)

print('RMSE =',RMSE_lr, '\nMSE =',MSE_lr, '\nMAE =',MAE_lr, '\nR2 =', r2_lr, '\nAdjusted R2 =', adj_r2_lr) 
```

# TASK 8 : TRAIN A XGBOOST REGRESSOR MODEL

```python
import xgboost as xgb
```

```python
xgb_model = xgb.XGBRegressor(objective='reg:squarederror')
```

```python
xgb_model.fit(X_train, y_train)
```

```python
xgb_model.score(X_test, y_test)
```

# TASK 9 : EVALUATE XGB TRAINED MODEL PERFORMANCE

```python
# Make prediction

y_predict_xgb = xgb_model.predict(X_test)
```

```python
# Plot the scaled result

plt.plot(y_test, y_predict_xgb, "^", color = 'r')
plt.xlabel('Model Predictions')
plt.ylabel('True Values')
```

```python
# Plot the KPIs

k = X_test.shape[1]
n = len(X_test)
RMSE_xgb = float(format(np.sqrt(mean_squared_error(y_test, y_predict_xgb)),'.3f'))
MSE_xgb = mean_squared_error(y_test, y_predict_xgb)
MAE_xgb = mean_absolute_error(y_test, y_predict_xgb)
r2_xgb = r2_score(y_test, y_predict_xgb)
adj_r2_xgb = 1-(1-r2_xgb)*(n-1)/(n-k-1)

print('RMSE =',RMSE_xgb, '\nMSE =',MSE_xgb, '\nMAE =',MAE_xgb, '\nR2 =', r2_xgb, '\nAdjusted R2 =', adj_r2_xgb) 
```

# TASK 10 : COMPARING BOTH TRAINED MODELS

```python
plt.scatter(y_test, y_predict_lr, label='Linear Regression', alpha=0.5)
plt.scatter(y_test, y_predict_xgb, label='XGBoost Regressor', alpha=0.5)
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.legend()
plt.title('Comparison of Linear Regression and XGBoost Regressor')
plt.show()
```

```python
import seaborn as sns

sns.kdeplot(y_predict_lr, label='Linear Regression Predictions', fill=True)
sns.kdeplot(y_test, label='Actual Values', fill=True)
sns.kdeplot(y_predict_xgb, label='XGBoost Regressor Predictions', fill=True)
plt.xlabel('Values')
plt.ylabel('Density')
plt.legend()
plt.title('Density Plot for Actual Values and Predictions')
plt.show()
```
